```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler 
from sklearn import datasets 
boston = datasets.load_boston()
features = pd.DataFrame(boston.data, columns=boston.feature_names)
targets = boston.target
```

```python
features
```

```python
features.describe()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 25), facecolor= 'white')
plotnumber = 1

for column in features:
    if plotnumber <=13:
        ax = plt.subplot(4,4 ,plotnumber)
        sns.distplot(features[column])
        plt.xlabel(column, fontsize=20)
    plotnumber+=1
plt.show()
```

```python
features.info()
```

```python
f,ax = plt.subplots(figsize=(20, 20))
sns.heatmap(features.corr(), annot=True) 
```

```python
Scaling = StandardScaler()
df = Scaling.fit_transform(features)
df = pd.DataFrame(df)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 25), facecolor= 'white')
plotnumber = 1

for column in df:
    if plotnumber <=13:
        ax = plt.subplot(4,4 ,plotnumber)
        sns.distplot(df[column])
        plt.xlabel(column, fontsize=20)
    plotnumber+=1
plt.show()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(df, targets, test_size=0.3)
```

```python
xtrain.shape, xtest.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=90)
rf_model.fit(xtrain, ytrain)
```

```python
rf_model.score(xtest, ytest)
```

```python
preds = rf_model.predict(xtest)
preds
```

```python
ytest
```

```python
from sklearn.metrics import mean_squared_error;
mean_squared_error(ytest, preds)
```
