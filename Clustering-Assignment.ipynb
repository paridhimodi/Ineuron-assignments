## In this assignment students have to compress racoon grey scale image into 5 clusters. In the end, visualize both raw and compressed image and look for quality difference.The raw image is available in spicy.misc package with the name face.

```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn import cluster, datasets
import scipy.misc
```

```python
f = scipy.misc.face(gray = True)
plt.figure(figsize=(10, 4))
plt.imshow(f, cmap = plt.cm.gray)
plt.show()
```

```python
f.shape
```

```python
row = f.shape[0]
col = f.shape[1]
image = f.reshape(row*col, 1)
image
```

```python
from sklearn.cluster import KMeans
k_mean = KMeans(n_clusters =5)
k_mean.fit(image)
```

```python
clusters = np.asarray(k_mean.cluster_centers_)
labels = np.asarray(k_mean.labels_)
labels = labels.reshape(row, col)
labels
```

```python
plt.imsave('compressed.png',labels, cmap=plt.cm.gray);
```

```python

plt.figure(1)
plt.imshow(f, cmap = plt.cm.gray)
plt.show()

image = plt.imread('compressed.png')
plt.figure(2)
plt.imshow(image)
plt.show()
```
